## Чтение файлов

### **.csv**

Формат `.csv` является самым распространенным при работе с Pandas.

Чтобы открыть файл формата `.csv` воспользуемся функцией `read_csv()`.

```python
# чтение из файла .csv
data = pd.read_csv('exmpl_csv.csv')
data
```

Если открыть файл `exmpl_csv.csv` через текстовый редактор (например блокнот), то мы увидим простую текстовую запись:

    number,name,kind
    1,goose,bird
    4,duck,bird
    9,bear,animal
    3,salmon,fish

Чтобы не создавать для каждого эксперимента новый текстовый файл для загрузки его через `read_csv()`, воспользуемся функцией `StringIO()` из библиотеки `io`. Модуль `StringIO` позволяет работать со строкой как с файловым объектом. Вот как это работает:

```python
from io import StringIO

# текст для чтения формата .csv
csvString = '''number,name,kind
               1,goose,bird
               4,duck,bird
               9,bear,animal
               3,salmon,fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO)
data
```

Структура стандартного `.csv` файла это построчная запись значений через запятую (разделитель). Количество разделителей в каждой строке может быть неодинаковым. В тех строках, в которых меньше разделителей будут добавляться значения `NaN` в конце.

```python
# текст для чтения формата .csv
csvString = '''number,name,kind
               1,bird
               4,duck,bird
               9,bear
               3,salmon,fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO)
data
```

Как мы видим, структура таких таблиц нарушается, поэтому лучше указывать одинаковое количество разделителей в каждой строке.

```python
# текст для чтения формата .csv
csvString = '''number,name,kind
              1,,bird
              4,duck,bird
              9,bear,
              3,salmon,fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO)
data
```

Часто, разделителем является знак точки с запятой `;`. Функция `read_csv()` не умеет автоматически определять разделитель.

```python
# текст для чтения формата .csv
csvString = '''number;name;kind
               1;goose;bird
               4;duck;bird
               9;bear;animal
               3;salmon;fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO)
data
```

Символ разделителя можно явно указать при помощи параметра `sep`.

```python
# текст для чтения формата .csv
csvString = '''number;name;kind
               1;goose;bird
               4;duck;bird
               9;bear;animal
               3;salmon;fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO, sep=';')
data
```

```python
# текст для чтения формата .csv
csvString = '''number&name&kind
               1&goose&bird
               4&duck&bird
               9&bear&animal
               3&salmon&fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO, sep='&')
data
```

Символ разделителя может быть любой, но чаще всего используются запятая `,` или точка с запятой `;`.

Если имена столбцов находятся не в первой строке, то их расположение можно указать явно при помощи параметра `header`. Но данные, которые находятся до этой строки включены не будут:

```python
# текст для чтения формата .csv
csvString = '''1;goose;bird
               4;duck;bird
               number;name;kind
               9;bear;animal
               3;salmon;fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO,
                   sep=';',
                   header=2)
data
```

По умолчанию имена столбцов берутся из 0 строки, поэтому если таблица не содержит имена, то это можно указать параметром `header=None`.

```python
# текст для чтения формата .csv
csvString = '''1,goose,bird
               4,duck,bird
               9,bear,animal
               3,salmon,fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO,
                   sep=',',
                   header=None)
data
```

Либо можно дать новые имена столбцам при помощи параметра `names`:

```python
# текст для чтения формата .csv
csvString = '''1,goose,bird
               4,duck,bird
               9,bear,animal
               3,salmon,fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO,
                   sep=',',
                   names=['Номер','Имя','Вид'])
data
```

В качестве индексов можно передать любой числовой столбец из таблицы в параметр `index_col`:

```python
# текст для чтения формата .csv
csvString = '''1,goose,bird
               4,duck,bird
               9,bear,animal
               3,salmon,fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO,
                   sep=',',
                   names=['Номер','Имя','Вид'],
                   index_col=0)
data
```

При считывании данных из файла мы можем указать тип данных каждого столбца:

```python
# текст для чтения формата .csv
csvString = '''number,name,kind
               1,goose,bird
               4,duck,bird
               9,bear,animal
               3,salmon,fish'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
data = pd.read_csv(csvStringIO,
                   sep=',',
                   dtype={0: float, 1: str, 2: str})
data
```

## Тип данных `category`

При работе с pandas мы часто используем типы данных `int`, `str`, `object`, `float`, но не многие знакомы с типом `category`. Разберём, для чего нужен категориальный тип данных и когда его использовать.

Категориальные данные - это данные, которые принимают конечное число возможных значений.

Преимущества использования категориальных данных:
 - компактный размер данных
 - возможность упорядочивания
 - поддержка построения графиков

Воспользуемся датасетом аквакультур, в котором выберем столбцы для перевода к категориальному типу.

```python
# база данных выращивания сомов и форели
data = pd.read_csv('https://query.data.world/s/7zlklw25pfhsquwzwyscqskjipok3w')
data
```

```python
# проверим тип данных датасета
data.info()
```

Выполним группировку по столбцу GEOGRAPHY_DESC и посчитаем сумму. Посчитаем сколько времени будет выполняться эта операция (потом сравним скорость выполнения с категориальным типом данных).

```python
%%timeit
data.groupby('GEOGRAPHY_DESC')['AMOUNT'].sum().to_frame()
```

Чтобы выбрать какие столбцы мы можем перевести в категориальный формат посчитаем количество уникальных значений по каждому из них. Для этого воспользуемся методом `.unique()`.

```python
# сколько уникальных значений в столбце SOURCE_ID
data.SOURCE_ID.nunique()
```

```python
# посчитаем количество уникальных значений для всех столбцов
for i in data.columns:
  print(i, data[i].nunique())
```

В таблице 20 тыс. строк, следовательно мы можем перевести все столбцы кроме AMOUNT к категориальномуу типу.

```python
# переведем выбранные столбцы к типу данных category
for i in data.columns:
  if i!='AMOUNT':
    data[i] = data[i].astype('category')
```

```python
# посмотрим инфо в обновленной таблице
data.info()
```

Изначальная таблица весила 1.4 Мб, обновленная стала 330 Кб. За счет перевода типа данных к категориям получилось значительно съэкономить память.

Использование категорий не только экономит память. Некоторые операции работают быстрее, например группировка `.groupby()`, которую мы выполнили ранее с подсчетом времени. Выполним эту операцию снова, но уже с категориальными данными:

```python
%%timeit
data.groupby('GEOGRAPHY_DESC')['AMOUNT'].sum().to_frame()
```

Время выполнения сократилось в 3 раза!

## Метод `.apply()`

```python
# создадим датафрейм для тестирования на нём метода apply()
csvString = '''month,shop1,shop2,shop3,shop4,shop5
               1,30,25,22,28,44
               2,32,27,23,33,42
               3,30,24,24,25,46
               4,35,35,25,23,50
               5,40,45,29,38,54
               6,43,43,32,41,64
               7,57,55,32,37,57
               8,46,35,22,36,53
               9,45,33,25,37,46
               10,37,28,24,28,43
               11,30,27,22,26,38
               12,28,25,21,28,37'''

# имитируем из текста текстовый файл
csvStringIO = StringIO(csvString)

# чтение из csvStringIO
df = pd.read_csv(csvStringIO, index_col=0)
df
```

Стандартный подход к изменению значений какого-либо столбца датафрейма обычно выглядит так:

```python
# добавим ко всем значениям столбца shop1 по 12
df.shop1 = df.shop1 + 12
df
```

Если мы хотим выполнить какую-то простую операцию для всех ячеек выбранного столбца, то такой подход вполне достаточен. Но что, если нам нужно выполнить математические операции с несколькими столбцами, да ещё и в зависимости от значения конкретной ячейки выполнять разные условия? (циклы не предлагать)

В таких случаях и используется метод `.apply()`.

В качестве параметра в apply передаётся функция (довольно часто лямбда функция).

```python
# вычтем из всех значений столбца shop2 по 20
df.shop2 = df.shop2.apply(lambda x: x-20)
df
```

Эквивалентом лямбда функции

    lambda x: x-20

является

    def func(x):
      return x-20

где x - это значение каждой ячейки.


```python
# выполним такую же операцию со столбцом shop3, но при помощи функции func
def func(x):
  return x-20

# указываем только имя функции
df.shop3 = df.shop3.apply(func)
df
```

Напишем функцию, которая прибавляет 100 к значениям больше или равным 30 и отнимает 10 из значений меньше 30.

```python
# функция
def hard_mat(x):
  if x >= 30:
    return x+100
  else:
    return x-10

# применим функцию hard_mat к столбцу shop4
df.shop4 = df.shop4.apply(hard_mat)
df
```

## Объединение таблиц

Подготовим несколько таблиц для занятия. Таблицы будут содержать данные интернет магазина.

```python
# таблица покупок 1-го магазина
shop_table = pd.DataFrame({'ID_Товара': ['0005', '0007', '0021', '0013'],
                          'ID_Покупателя': ['22074', '22069', '22074', '22056'],
                          'Количество': [1, 4, 3, 2],
                          'Статус': ['Оплачен', 'Отменен', 'Оплачен', 'Ожидает оплаты']})

# таблица покупок 2-го магазина
shop_table2 = pd.DataFrame({'ID_Товара': ['1032', '0432', '0005', '0021'],
                            'ID_Покупателя': ['21000', '21018', '21019', '21020'],
                            'Количество': [10, 2, 1, 2],
                            'Статус': ['Оплачен', 'Оплачен', 'Ожидает оплаты', 'Оплачен']})

# таблица товаров на складе
product_table = pd.DataFrame({'Наименование': ['Тетрадь', 'Дневник', 'Ручка', 'Ластик', 'Пенал', 'Набор карандашей'],
                              'ID_Товара': ['1032', '0021', '0005', '0013', '0432', '0007'],
                              'Скидка': ['3%', '5%', '3%', '10%', '10%', '5%',],
                              'Остаток': [543, 434, 1043, 233, 543, 761]})
```

### Объединение таблиц по строкам

####`pd.concat()`

Объединим таблицы покупок обоих магазинов. Обе таблицы имеют идентичные колонки, поэтому мы можем в одну таблицу добавить все строки из другой таблицы.

```python
# воспользуемся функцией concat
pd.concat((shop_table, shop_table2))
```

Строки таблицы shop_table2 добавились в конец таблицы shop_table. Если мы поменяем порядок аргументов функции concat, то первая таблица будет добавлена в конец второй.

```python
# поменяем аргументы местами
pd.concat((shop_table2, shop_table))
```

При таком объединении индексы остались прежними, поэтому, если мы хотим их обновить, то нужно включить параметр `ignore_index=True`.

```python
# обновляем индексы
pd.concat((shop_table2, shop_table), ignore_index=True)
```

Функция concat работает по любой оси - как по строкам, так и по столбцам. По какой оси делать объединение можно указать при помощи параметра `axis`. По умолчанию этот параметр 0, т.е. делает объединение по строкам.

Попробуем объединить наши таблицы по столбцам.

```python
pd.concat((shop_table, shop_table2), axis=1)
```

Таблица просто добавляется справа, увеличивая общее количество столбцов.

####`df.append()`

Так же, для объединения таблиц по строкам есть метод `append`, который так же добавляет строки одной таблицы в конец другой.

```python
shop_table.append(shop_table2, ignore_index=True)
```

**CONCAT vs APPEND**

*   **Concat** дает возможность объединения на основе оси (по строкам или по столбцам)
*   **Append** - это частный случай concat (только ось 0). Считается устаревшей механикой.

### Объединение таблиц по столбцам

Теперь давайте попробуем объединить таблицы shop_table и product_table. Это разные таблицы, поэтому по строкам их объединять не имеет смысла, но можно заметить, что таблица product_table имеет расширенную информацию о товарах, поэтому мы можем для каждой покупки (строки) таблицы shop_table добавить информацию о выбранном товаре.

```python
# объединим таблицы по оси 1
pd.concat((shop_table, product_table), axis=1)
```

Казалось бы, объединение прошло успешно, но во-первых, у нас появились значения NaN, а во-вторых, `ID_Товара` теперь 2 столбца и они не идентичны. Так как столбцы `ID_Товара` не одинаковы, следовательно и характеристики самих товаров не соответствуют покупкам.

Вывод: `pd.concat()` в данном случае не подходит для объединения таблиц.

####`pd.merge()`

Для объединения таблиц, у которых не все столбцы идентичны используется функция `merge()`.

Если мы посмотрим на таблицы shop_table и product_table, то заметим, что у них есть одинаковый столбец `ID_Товара`. На основе этого столбца можно выполнить объединение, а такой столбец принято называть **ключом**.

Объединим таблицы с параметром how='left'.

image.png

```python
# объединение таблиц по ключу
pd.merge(shop_table, product_table, how='left', on='ID_Товара')
```

Параметры функции:

    pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None, left_index=False, right_index=False)

---
    left : DataFrame или Series
          Левая таблица.
    
    right : DataFrame или Series
          Правая таблица.

    how : {‘left’, ‘right’, ‘outer’, ‘inner’}, по умоляанию ‘inner’
          Метод объединения
          *  'left' : используются ключи только левой таблицы
          *  'right' : используются ключи только правой таблицы
          *  'inner' : используются ключи, которые есть только в обоих таблицах (пересечение)
          *  'outer' : используются все ключи, обоих таблиц (объединение)

    on : label или list
          Имя столбца или индекс, который будет использован в качестве ключа.
          Должен иметь одинаковое имя в обоих таблицах.

    left_on : label или list
          Имя столбца или индекс, который будет использован в качестве ключа левой таблицы.

    right_on : label или list
          Имя столбца или индекс, который будет использован в качестве ключа правой таблицы.

    left_index : bool, по умолчанию False
          Использовать ли индекс левой таблицы в качестве ключа.

    right_index : bool, по умолчанию False
          Использовать ли индекс правой таблицы в качестве ключа.


Используя параметр `how='left'` мы получили все строки левой таблицы без изменений + соответствующие совпавшим ключам значения из правой таблицы.

Объединим таблицы с параметром `how='right'`.

image.png

```python
# объединение таблиц по ключу
pd.merge(shop_table, product_table, how='right', on='ID_Товара')
```

Теперь к строкам правой таблицы product_table добавились соответствующие совпавшим ключам значения левой таблицы shop_table. Значения NaN появились в тех ячейках, которым не нашлось пары совпадений по ключу в обоих таблицах.

Объединим таблицы с параметром how='inner'.

image.png

```python
# объединение таблиц по ключу
pd.merge(shop_table, product_table, how='inner', on='ID_Товара')
```

Объединим таблицы с параметром how='outer'.

image.png

```python
# объединение таблиц по ключу
pd.merge(shop_table, product_table, how='outer', on='ID_Товара')
```

Бывает так, что имена столбцов, которые является ключами в разных таблицах не одинаковы, тогда вместо параметра `on` нужно использовать параметры `left_on` и `right_on`.

```python
# поменяем имя столбца ID_Товара в таблице product_table
product_table.rename(columns = {'ID_Товара':'prod_ID'}, inplace = True )
```

```python
# проверим изменения
product_table
```

Объединим таблицы со старым параметром on.

```python
# объединение таблиц по ключу
pd.merge(shop_table, product_table, how='left', on='ID_Товара')
```

Ошибка! Так как указанное имя ключа не найдено в одной из таблиц.

Воспользуемся параметрами `left_on` и `right_on`.

```python
# объединение таблиц по ключу
pd.merge(shop_table, product_table, how='left', left_on='ID_Товара', right_on='prod_ID')
```

По значениям столбцов ID_Товара и prod_ID видно, что объединение выполнено без ошибок.

####`df.join()`

Метод `join()` можно считать упрощенной версией функции `merge()`. Он имеет меньше функционала и вот его основные параметры:

    DataFrame.join(other, on=None, how='left')

---

    other : DataFrame или Series
          Правая таблицы объединения.

    on : label или list
          Имя столбца или индекс, который будет использован в качестве ключа.
          Должен иметь одинаковое имя в обоих таблицах.

    how : {‘left’, ‘right’, ‘outer’, ‘inner’}, по умоляанию ‘inner’
          Метод объединения
          *  'left' : используются ключи только левой таблицы
          *  'right' : используются ключи только правой таблицы
          *  'inner' : используются ключи, которые есть только в обоих таблицах (пересечение)
          *  'outer' : используются все ключи, обоих таблиц (объединение)

Объединение возможно либо по индексам, либо по ключу с одинаковым именем. Если не указывать ключ `on`, то объединение будет по индексам.

```python
product_table
```

```python
# объединение таблиц по ключу (по умолчанию ключом являются индексы)
shop_table.join(product_table, how='left')
```

**MERGE vs JOIN**

*   **Merge** дает возможность объединения по ключу (как с одинаковыми именами, так и с разными), индексам и в разных комбинациях.
*   **Join** дает возможность объединения по ключу (только с одинаковыми именами) и по индексам.
